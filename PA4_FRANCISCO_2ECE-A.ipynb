```python
import pandas as pd
```

```python
#READ THE FILE
ECE = pd.read_excel('board2.xlsx')
ECE
```

# PROBLEM 1: ECE BOARD EXAM PROBLEM

```python
#1.a
#ASSIGN THE LOC TO INSTRU TO USE TO SHOW THE DATA / CREATES THE OLY SPECIFIED DATA AND GRAPH NEEDED
Instru = ECE.loc[
    (ECE['Track'] == 'Instrumentation')&
    (ECE['Hometown'] == 'Luzon')&
    (ECE['Electronics'] > 70 ),
    ['Name', 'GEAS', 'Electronics']
]
Instru
```

```python
#1.a
#PUT ALL THE COURSES IN ONE SUBJECT, FUNCTION TO CALL ALL OF THEM ONE NEEDED TO GET A SPECIFIC COMPUTATION
subjects = ['Math', 'Electronics', 'GEAS', 'Communication']
#COMPUTES THE MEAN
ECE['Average'] = ECE[subjects].mean(axis=1)

Mindy = ECE.loc[
    (ECE['Hometown'] == 'Mindanao') &
    (ECE['Gender'] == 'Female') &
    (ECE['Average'] >= 55),
    ['Name', 'Track', 'Electronics', 'Average']
]
Mindy
```

# PROBLEM 2: VISUALIZATION FOR SPECIFIC HIGHER AVERAGE

```python
import matplotlib.pyplot as plt
```

```python
#MEAN COVERTION FOR AVERAGE OF THE SPECIFIED COLUMN OF GENDER
gaverage1 = ECE.groupby('Gender')['Average'].mean()

#FORMAT OF THE GRAPH AS OF WHAT NEEDED FOR GRAPHING
plt.figure(figsize=(4,4))
plt.bar(gaverage1.index, gaverage1.values)
plt.title('Average Comparison (Gender)')
plt.xlabel('Gender')
plt.ylabel('Average Grade')
plt.show()
```

```python
#MEAN COVERTION FOR AVERAGE OF THE SPECIFIED COLUMN OF HOMETOWN
gaverage2 = ECE.groupby('Hometown')['Average'].mean()

#FORMAT OF THE GRAPH AS OF WHAT NEEDED FOR GRAPHING
plt.figure(figsize=(6,4))
plt.bar(gaverage2.index, gaverage2.values)
plt.title('Average Comparison (Geographical Divisions)')
plt.xlabel('Hometown')
plt.ylabel('Average Grade')
plt.show()
```

```python
#MEAN COVERTION FOR AVERAGE OF THE SPECIFIED COLUMN OF TRACK
gaverage3 = ECE.groupby('Track')['Average'].mean()

#FORMAT OF THE GRAPH AS OF WHAT NEEDED FOR GRAPHING
plt.figure(figsize=(6,4))
plt.bar(gaverage3.index, gaverage3.values)
plt.title('Average Comparison (Subjects)')
plt.xlabel('Track')
plt.ylabel('Average Grade')
plt.show()
```
